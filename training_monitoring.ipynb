```python
import json
import matplotlib.pyplot as plt
from detectree2.models.train import load_json_arr

out_dir = "/mnt/e_drive/detectree2/griffith_site/train_outputs_allsites"
experiment_folder = out_dir

experiment_metrics = load_json_arr(experiment_folder + '/metrics.json')

# sort out the metrics data

mask_rcnn_acc = ([x['iteration'] for x in experiment_metrics if 'mask_rcnn/accuracy' in x], 
                 [x['mask_rcnn/accuracy'] for x in experiment_metrics if 'mask_rcnn/accuracy' in x], 'segmentation acc Vs iteration')
Mask_loss = ([x['iteration'] for x in experiment_metrics if 'loss_mask' in x],
            [x['loss_mask'] for x in experiment_metrics if 'loss_mask' in x], 'Segmentation loss Vs Iteration')
False_negative_Masks = ([x['iteration'] for x in experiment_metrics if 'mask_rcnn/false_negative' in x],
                    [x['mask_rcnn/false_negative'] for x in experiment_metrics if 'mask_rcnn/false_negative' in x], 'segmentation false negative Vs Iteration')
False_positive_Masks = ([x['iteration'] for x in experiment_metrics if 'mask_rcnn/false_positive' in x],
                    [x['mask_rcnn/false_positive'] for x in experiment_metrics if 'mask_rcnn/false_positive' in x], 'Segmentation false positive Vs Iteration')
learning_rate =  ([x['iteration'] for x in experiment_metrics if 'lr' in x],
                    [x['lr'] for x in experiment_metrics if 'lr' in x], 'Learning rate Vs Iteration')

y = [mask_rcnn_acc, Mask_loss, False_positive_Masks, learning_rate ]

fig, axes = plt.subplots(2, 2, figsize=(8, 8))

for i, axes in enumerate(axes.flat):
    axes.set_title(y[i][2])
    axes.plot(y[i][0],y[i][1])  # Example plot
    axes.grid(True)
 
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
    
    
```
